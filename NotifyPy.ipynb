```python
import requests
from bs4 import BeautifulSoup
import time
import os
from dotenv import load_dotenv
```

```python
load_dotenv()
URL = os.getenv("URL")
API_ENDPOINT = os.getenv("API_ENDPOINT")
CHAT_ID = os.getenv("CHAT_ID")
TAG_ID = os.getenv("TAG_ID")

HEADERS = {"User-Agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/83.0.4103.116 Safari/537.36"}
API_ENDPOINT_PARAMS = {
    'chat_id': CHAT_ID, 
    'text': URL + " is in stock!"
}

def get_page_html(url):
    page = requests.get(url, headers=HEADERS)
    print(page.status_code)
    return page.content


def check_item_in_stock():
    page_html = get_page_html(URL)
    print(URL)
    soup = BeautifulSoup(page_html, 'html.parser')
    buyButtonText = soup.find(id=TAG_ID).getText().strip()
    print(buyButtonText)
    return "Sold Out" not in buyButtonText


def notify():
    requests.post(url = API_ENDPOINT, params=API_ENDPOINT_PARAMS) 
```

```python
while 1:
    if check_item_in_stock():
        time.sleep(2)
        notify() 
```
